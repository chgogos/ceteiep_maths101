```python
from sympy import Symbol
from sympy.plotting import plot

x= Symbol('x')
expr = 2 
p=plot(expr, (x,-10,10), title= expr)
```

```python
from sympy import Symbol
from sympy.plotting import plot

x= Symbol('x')
expr = x
p=plot(expr, (x,-10,10), title= expr)
```

```python
from sympy import Symbol
from sympy.plotting import plot

x= Symbol('x')
expr = x**2
p=plot(expr, (x,-10,10), title= expr)
```

```python
from sympy import Symbol
from sympy.plotting import plot

x= Symbol('x')
expr = x**3
p=plot(expr, (x,-10,10), title= expr)
```

```python
from sympy import Symbol
from sympy.plotting import plot

x= Symbol('x')
expr = 1/x
p=plot(expr, (x,-10,10), title= expr, ylim=(-10,10))
```

```python
from sympy import Symbol
from sympy.plotting import plot

x= Symbol('x')
expr = abs(x)
p=plot(expr, (x,-10,10), title= expr)
```

```python
from sympy import Symbol
from sympy.plotting import plot

x= Symbol('x')
expr = 2**x
p=plot(expr, (x,-10,10), title= expr)
```

```python
from sympy import Symbol, sin
from sympy.plotting import plot

x= Symbol('x')
expr = sin(x)
p=plot(expr, (x,-10,10), title= expr)
```

```python
from sympy import Symbol, cos, pi
from sympy.plotting import plot

x= Symbol('x')
expr = cos(x)
p=plot(expr, (x,-2*pi,2*pi), title= expr)
```

```python
from sympy import Symbol, exp
from sympy.plotting import plot

x= Symbol('x')
expr = exp(x)
p=plot(expr, (x,-10,10), title= expr, ylim=(0,100))
```

```python
from sympy import Symbol, ln
from sympy.plotting import plot

x= Symbol('x')
expr = ln(x)
p=plot(expr, (x,-10,10), title= expr)
```
