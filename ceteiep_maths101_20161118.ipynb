####  Γραφικές παραστάσεις συναρτήσεων - μετασχηματισμοί συναρτήσεων

```python
%matplotlib inline
from sympy import *
from sympy.plotting import plot
init_printing()

x= Symbol('x')
```

```python
# ταυτοτική συνάρτηση
expr = x 
plot(expr, (x,-10,10),ylim=(-10,10), title= expr);
```

```python
# συνάρτηση 1 προς x
expr = 1/x 
plot(expr, (x,-10,10),ylim=(-10,10), title= expr);
```

```python
# συνάρτηση απόλυτης τιμής
expr = abs(x)
plot(expr, (x,-10,10), title= expr);
```

```python
# τετραγωνική συνάρτηση
expr = x**2 
plot(expr, (x,-10,10),ylim=(0,20), title= expr);
```

```python
# κυβική συνάρτηση
expr = x**3 
plot(expr, (x,-10,10),ylim=(-20,20), title= expr);
```

```python
# συνάρτηση τετραγωνικής ρίζας
expr = sqrt(x) 
plot(expr, (x,0,10),ylim=(0,5), title= expr)
```

```python
# εκθετική συνάρτηση
# expr = E**x
expr = exp(x)
plot(expr, (x,-10,10),ylim=(0,20), title= expr);
```

```python
# λογαριθμική συν΄άρτηση
expr = log(x) 
plot(expr, (x,-10,10),ylim=(-5,5), title= expr);
```

```python
# συνάρτηση ημιτόνου
expr = sin(x) 
plot(expr, (x,-10,10),ylim=(-1,1), title= expr);
```

```python
# συνάρτηση συνημιτόνου
expr = cos(x) 
plot(expr, (x,-10,10),ylim=(-1,1), title= expr);
```

#### Μετασχηματισμός της συνάρτησης $f(x)=x^3-x-2$
1. Δεξιά μετατόπιση κατά 2
2. Κατακόρυφη μετατόπιση κατά 5 προς τα κάτω 
3. Συμμετρία ως προς τον κατακόρυφο άξονα

Προκύπτει η συνάρτηση
$$f(x) = (-x-2)^3 - (-x - 2) - 2 - 5 = -x^3-6x^2-11x-13$$ 

```python
# expr = x**3 - x - 2 
# expr = (x-2)**3 - (x - 2) - 2  # μετακίνηση προς τα δεξιά κατά 2
# expr = (x-2)**3 - (x - 2) - 2 - 5 # μετακίνηση προς τα κάτω κατά 5
# expr = (-x-2)**3 - (-x - 2) - 2 - 5 # οριζόντια αντιστροφή
expr = (-x-2)**3 + x - 5
plot(expr, (x,-10,10),ylim=(-20,20))
pprint(expand(expr))
```

```python
# εναλλακτικά 
from IPython.display import display
expr = x**3 - x - 2 
expr = expr.subs(x, x-2)
expr = expr - 5
expr = expr.subs(x,-x)
plot(expr, (x,-10,10),ylim=(-20,20))
display(expand(expr))
```

#### Σύνθεση συναρτήσεων
Ο όγκος μιας σφαίρας σε σχέση με την ακτίνα του $r$ είναι $V=\frac{4}{3}\pi r^3$. Έστω ότι η ακτίνα της σφαίρας μεγαλώνει συναρτήσει του χρόνου $t$ σε δευτερόλεπτα σύμφωνα με την σχέση $r(t)=3\sqrt(t)$. 
* Ποια θα είναι η συνάρτηση του όγκου της σφαίρας σε σχέση με το χρόνο; 
* Τι όγκο θα έχει η σφαίρα μετά από 1 λεπτό;


```python
r, t = symbols('r t')
v_r = 4/3*pi*r**3
r_t = 3*sqrt(t)
f = v_r.subs(r, r_t)
display(f)
f.subs(t,60).evalf() # τιμή συνάρτησης για t=60
```
