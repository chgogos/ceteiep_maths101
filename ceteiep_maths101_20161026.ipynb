```python
from sympy import *
from IPython.display import display
from IPython.display import Latex 
init_printing()
```

Γραμμικός μετασχηματισμός των σημείων (1,2) και (1,-2) με πίνακα γραμμικού μετασχηματισμού τον $\begin{bmatrix} 7 & 3 \\ 2 & 2 \end{bmatrix}$

```python
T = Matrix([[7,3],[2,2]]) # πίνακας γραμμικού μετασχηματισμού
display(T)
A = Matrix([[1],[2]]) 
Aprime = T*A
display(Latex("$A=%s A\'=%s$" % (latex(A), latex(Aprime))))
B = Matrix([[1],[-2]])
Bprime = T*B
display(Latex("$B=%s B\'=%s$" % (latex(B), latex(Bprime))))
```

```python
# προκειμένου να απεικονίζεται το γράφημα στο notebook θα πρέπει να δοθεί η ακόλουθη εντολή
%matplotlib inline  
import matplotlib.pyplot as plt 
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) # 1x1 πλέγμα υπογραφημάτων, απεικόνιση στο 1 (και μοναδικό) υπογράφημα
plt.xlim([-1,14]) # όρια x άξονα (από-έως)
plt.ylim([-3,7]) # όρια y άξονα (από-έως)
plt.axhline(0, color='black') # θέση στην οποία θα ζωγραφιστεί οριζόντια γραμμή και χρώμα
plt.axvline(0, color='black') # θέση στην οποία θα ζωγραφιστεί κατακόρυφη γραμμή και χρώμα
plt.grid('on'); # εμφάνιση πλέγματος
# point1 = (1,2)  --> (13,6)
# point2 = (1,-2) --> (1,-2)
plt.plot([1,1],[2,-2],'ro',  ms=20) # απεικόνιση των σημείων (1,2) και (1,-2) με κόκκινο χρώμα και σχήμα κύκλου
plt.plot([13,1],[6,-2],'b*', ms=20) # απεικόνιση των σημείων (13,6) και (1,-2) με μπλε χρώμα και σχήμα αστεριού
ax.annotate('A', xy=(1,2), xytext=(4, 4),
            arrowprops=dict(facecolor='black', shrink=0.1)) # υποσημείωση στο γράφημα (κείμενο και βέλος)
ax.annotate('B', xy=(1,-2), xytext=(4, -0.5),
            arrowprops=dict(facecolor='black', shrink=0.1))
ax.annotate("A'", xy=(13,6), xytext=(10,5),
            arrowprops=dict(facecolor='black', shrink=0.1))
ax.annotate("B'", xy=(1,-2), xytext=(4, -2),
            arrowprops=dict(facecolor='black', shrink=0.1));
```

#### Γραμμικός μετασχηματισμός πολυγώνου
Έστω ένα τρίγωνο το οποίο ορίζεται από τα σημεία (1,0), (-1,3) και (0,3). Να παρουσιαστεί ο γραμμικός μετασχηματισμός του με βάση τον πίνακα γραμμικού μετασχηματισμού $\begin{bmatrix} 2 & 1 \\ -1 & 0 \end{bmatrix}$

```python
A = Matrix([[1,-1,0],[0,3,3]])
T = Matrix([[2,1],[-1,0]])
AT = T*A
display(Latex("$A=%s A\'=%s$" % (latex(A), latex(AT))))
```

```python
from matplotlib.path import Path
import matplotlib.patches as patches
verts = [
    (1, 0),  # P0
    (-1, 3), # P1
    (0, 3), # P2
    (1, 0),  # P0
    ]
verts2 = [
    (2, -1),  # P0
    (1, 1), # P1
    (3, 0), # P2
    (2, -1),  # P0
    ]

codes = [Path.MOVETO,
         Path.LINETO,
         Path.LINETO,
         Path.CLOSEPOLY
         ]
path = Path(verts, codes)
path2 = Path(verts2, codes)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
patch = patches.PathPatch(path, facecolor='orange', linewidth=2)
ax.add_patch(patch)
ax.annotate('A', xy=(1,0))
ax.annotate('B', xy=(-1,3))
ax.annotate('C', xy=(0,3))
patch = patches.PathPatch(path2, facecolor='cyan', linewidth=2)
ax.add_patch(patch)
ax.annotate("A'", xy=(2,-1), xytext=(2,-1.2))
ax.annotate("B'", xy=(1,1))
ax.annotate("C'", xy=(3,0))
ax.set_aspect('equal')
ax.set_xlim(-1.5, 3.5)
ax.set_ylim(-1.5, 3.5)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title('linear transformation')

plt.grid();
```

#### Βασικοί γεωμετρικοί μετασχηματισμοί
* Συμμετρία ως προς την αρχή των αξόνων
* Συμμετρία ως προς τον άξονα των τετμημένων (άξονας x)
* Συμμετρία ως προς τον άξονα των τεταγμένων (άξονας y)
* Συμμετρία ως προς την ευθεία y=x
* Στροφή με κέντρο την αρχή των αξόνων και γωνία θ
* Ομοιοθεσία (scaling) με κέντρο την αρχή των αξόνων και λόγο λ
* Παράλληλη μεταφορά

```python
A = Matrix([[2,4,3,1],[1,3,5,2]])

# Συμμετρία ως προς την αρχή των αξόνων
T = Matrix([[-1,0],[0,-1]])
AT = T*A
display(Latex("$A=%s T=%s$ συμμετρία ως προς την αρχή των αξόνων $A\'=%s$" % (latex(A), latex(T), latex(AT))))

# Συμμετρία ως προς τον άξονα των τετμημένων (άξονας x)
T = Matrix([[1,0],[0,-1]])
AT = T*A
display(Latex("$A=%s T=%s$ συμμετρία ως προς τον άξονα των τετμημένων (άξονας x) $A\'=%s$" % (latex(A), latex(T), latex(AT))))

# Συμμετρία ως προς τον άξονα των τεταγμένων (άξονας y)
T = Matrix([[-1,0],[0,1]])
AT = T*A
display(Latex("$A=%s T=%s$ συμμετρία ως προς τον άξονα των τεταγμένων (άξονας y) $A\'=%s$" % (latex(A), latex(T), latex(AT))))

# Συμμετρία ως προς την ευθεία y=x
T = Matrix([[0,1],[1,0]])
AT = T*A
display(Latex("$A=%s T=%s$ συμμετρία ως προς την ευθεία y=x $A\'=%s$" % (latex(A), latex(T), latex(AT))))

# Στροφή με κέντρο την αρχή των αξόνων και γωνία θ
theta = pi/2
T = Matrix([[cos(theta),-sin(theta)],[sin(theta),cos(theta)]])
AT = T*A
display(Latex("$A=%s T=%s$ στροφή με κέντρο την αρχή των αξόνων και γωνία θ$=90^ο$ $A\'=%s$" % (latex(A), latex(T), latex(AT))))

# Ομοιοθεσία με κέντρο την αρχή των αξόνων και λόγο λ
l = 0.5
T = Matrix([[l,0],[0,l]])
AT = T*A
display(Latex("$A=%s T=%s$ Ομοιοθεσία με κέντρο την αρχή των αξόνων και λόγο λ=0.5 $A\'=%s$" % (latex(A), latex(T), latex(AT))))

# Παράλληλη μεταφορά 3 αριστερά και 3 προς τα κάτω
T = Matrix([[1,0],[0,1]])
B = Matrix([1,-2])
BX4 = B * ones(1,4)
AT = T*A + BX4
display(Latex("$A=%s T=%s $ Παράλληλη μεταφορά 1 δεξιά και 2 προς τα κάτω $%s A\'=%s$" % (latex(A), latex(T), latex(B), latex(AT))))
```
